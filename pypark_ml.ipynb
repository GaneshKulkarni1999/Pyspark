```python
from pyspark.ml.feature import VectorAssembler
featureassember=VectorAssembler(inputCols=['Rating','Salary'],outputCol='Independent')
```

```python
output = featureassember.transform(training)
output.show()
```

```python
finalized_data=output.select("Company Name","Salary")
```

```python
finalized_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
#training 
train_data,test_data=finalized_data.randomSplit([0.75,0.25])
regressor=LinearRegression(featureCol='Independent',labelCol='Salary')
regressor=regressor.fit(train_data)
```
